```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, StratifiedKFold
from scipy.stats import wilcoxon



print()
print("*****Breast Cancer prediction*****")
print()
breast_df = pd.read_csv("C:\\Nishu\\breast-cancer.csv")
breast_df.head()

breast_df.shape

breast_df.info()

breast_df.describe()

breast_df.isnull().sum()

breast_df['diagnosis'].value_counts() 

breast_df['diagnosis'] = breast_df['diagnosis'].astype('category')
breast_df.dtypes

#creating object of LabelEncoder 
le_y = LabelEncoder()
breast_df.iloc[:,1] = le_y.fit_transform(breast_df.iloc[:,1].values)

breast_df['diagnosis']

breast_df.groupby('diagnosis').mean()

# sns.pairplot(df.iloc[:,1:6], hue = 'diagnosis')

breast_df

breast_x = breast_df.drop('diagnosis', axis = 1)
breast_y = np.array(breast_df['diagnosis'])

#appling the feature selection model to the dataset
etc = ExtraTreesClassifier()
etc.fit(breast_x, breast_y)

print(etc.feature_importances_)

feature_importance = pd.Series(etc.feature_importances_, index = breast_x.columns)
feature_importance.nlargest(12).plot(kind = 'barh')
plt.show()

# corr = df.corr()
# top_feat = corr.index
# plt.figure(figsize = (20,20))

# heat = sns.heatmap(df[top_feat].corr(), annot = True )

breast_feat_corr =breast_df[['perimeter_mean', 'perimeter_worst','area_mean','area_worst','radius_mean','radius_worst','concave points_worst','compactness_mean','area_se', 'concave points_mean','concavity_mean','concavity_worst']].corr()
plt.figure(figsize = (15,10))
sns.heatmap(breast_feat_corr, annot = True)

breast_feat =breast_df[['perimeter_mean', 'perimeter_worst','area_mean','area_worst','radius_mean','radius_worst','concave points_worst','compactness_mean','area_se', 'concave points_mean','concavity_mean','concavity_worst']]

# observinng outliers
plt.figure(figsize = (12,12))

for i , col in enumerate(breast_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =breast_feat)
    
plt.show()



plt.figure(figsize = (12,12))
for i , col in enumerate(breast_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.histplot(x =  col, data = breast_feat, kde= True)
    
plt.show()

from scipy import stats
plt.figure(figsize = (12,12))
for i , col in enumerate(breast_feat.columns[:12]):
    plt.subplot(6,2,i+1)
    stats.probplot(breast_feat[col], dist = 'norm', plot = plt)
    
plt.show()

new_breast_feat = breast_feat.copy()
def outliers(df):
    # outlier detection and removal
    def detect_outliers_IQR(df, feature):
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return lower_bound, upper_bound
    
    # Define a function to cap outliers
    def cap_outliers(df, feature, lower_bound, upper_bound):
        df[feature] = np.where(df[feature] < lower_bound, lower_bound, df[feature])
        df[feature] = np.where(df[feature] > upper_bound, upper_bound, df[feature])
        return df
    
    # Loop through each feature in the DataFrame
    for feature in df.columns:
        # Detect outliers using IQR method
        lower_bound, upper_bound = detect_outliers_IQR(df, feature)
        
        # Cap outliers
        clean_df = cap_outliers(df, feature, lower_bound, upper_bound)
    
    # Print the resulting DataFrame
    return clean_df

breast_clean_df = outliers(new_breast_feat)

plt.figure(figsize = (12,12))

for i , col in enumerate(breast_clean_df.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =breast_clean_df)
    
plt.show()

x_train, x_test, y_train, y_test = train_test_split(breast_clean_df, breast_y,
                                                    test_size = 0.25, random_state = 0)

sc = StandardScaler()

breast_x_train = sc.fit_transform(x_train)
breast_x_test = sc.fit_transform(x_test)

#create a function for the models
def models(x_train, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    breast_log = LogisticRegression(random_state = 0)
    breast_log.fit(breast_x_train, y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    breast_tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    breast_tree.fit(breast_x_train, y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    breast_forest = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state = 0)
    breast_forest.fit(breast_x_train, y_train)

    #Support Vector Classifier
    from sklearn.svm import SVC
    breast_svc = SVC(kernel = 'rbf', random_state = 0, probability = True)
    breast_svc.fit(breast_x_train, y_train)

    # print the model accuracy on the training data
    print('[0] Logistic Regression Training Accuracy: ', breast_log.score(breast_x_train , y_train))
    print('[1] Decision Tree Classifier Training Accuracy: ', breast_tree.score(breast_x_train , y_train))
    print('[2] Random Forest Classifier Training Accuracy: ', breast_forest.score(breast_x_train , y_train))
    print('[3] Support Vector Classifier Training Accuracy: ', breast_svc.score(breast_x_train , y_train))

    return breast_log, breast_tree, breast_forest, breast_svc

model = models(breast_x_train, y_train)


# function for the cross-validation of models w.r.t their accuracy scoring
def cross_validation_evaluation(models, X, y, cv=5):
    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=0)
    model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVC']
    
    for model, name in zip(models, model_names):
        scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
        print(f'{name} Cross-Validation Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})')

# Evaluate models with cross-validation
cross_validation_evaluation(model, breast_clean_df, breast_y)

#test model accuracy on test data on cofusion matrix
from sklearn. metrics import confusion_matrix

print('0 - Logistic Regression \n1 - Decision Tree Classifier \n2 - Random Forest Classifier \n3 - SVC')

for i in range(len(model)):
    print(model[i])
    cm = confusion_matrix(y_test, model[i].predict(breast_x_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]

    print(cm)
    print('Testing Accuracy = ', (TP+TN)/(TP+TN+FP+FN))
    print()

    if TN + FP == 0:
        print('0')
    else:
         print('Specificity - ',TN / (TN + FP))


from sklearn.metrics import roc_curve, auc

breast_log_pred = model[0].decision_function(breast_x_test)
breast_dcn_tree_pred = model[1].predict_proba(breast_x_test)[:,1]
breast_rndm_forest_pred = model[2].predict_proba(breast_x_test)[:,1]
breast_svc_pred =  model[3].decision_function(breast_x_test)


breast_log_fpr, breast_log_tpr, threshold = roc_curve(y_test, breast_log_pred)
breast_auc_log = auc(breast_log_fpr, breast_log_tpr)

breast_svc_fpr, breast_svc_tpr, threshold = roc_curve(y_test, breast_svc_pred)
breast_auc_svc = auc(breast_svc_fpr, breast_svc_tpr)

breast_decision_tree_fpr, breast_decision_tree_tpr, threshold = roc_curve(y_test, breast_dcn_tree_pred)
breast_auc_dcn_tree = auc(breast_decision_tree_fpr, breast_decision_tree_tpr)

breast_rndm_forest_fpr, breast_rndm_forest_tpr, threshold = roc_curve(y_test, breast_rndm_forest_pred)
breast_auc_rndm_forest = auc(breast_rndm_forest_fpr, breast_rndm_forest_tpr)


print(breast_auc_log)
print(breast_auc_dcn_tree)
print(breast_auc_rndm_forest)
print(breast_auc_svc)

plt.figure(figsize = (5,5), dpi = 100)
plt.plot(breast_log_fpr, breast_log_tpr, linestyle = '-', label = 'Logistic Regression (auc = %0.3f)' %breast_auc_log)

plt.plot(breast_decision_tree_fpr, breast_decision_tree_tpr, linestyle= '-', label = 'Decision Tree (auc = %0.3f)' %breast_auc_dcn_tree)

plt.plot(breast_rndm_forest_fpr, breast_rndm_forest_tpr, linestyle = '-', label = 'Random Forest Classifier (auc = %0.3f)' %breast_auc_rndm_forest)

plt.plot(breast_svc_fpr, breast_svc_tpr, linestyle = '-', label = 'SVC (auc = %0.3f)' %breast_auc_svc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()

#classification report 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print(model[i])
    print(classification_report(y_test, model[i].predict(breast_x_test)))
    print(accuracy_score(y_test, model[i].predict(breast_x_test)))
    print()




def wilcoxon_signed_rank_test(model1, model2, X_test, y_test):
    # Predict probabilities
    y_pred1 = model1.predict_proba(X_test)[:, 1]
    y_pred2 = model2.predict_proba(X_test)[:, 1]
    differences = y_pred1 - y_pred2
    
    # Perform Wilcoxon signed-rank test
    statistic, p_value = wilcoxon(differences)
    
    mean_difference = np.mean(differences)
    
    # Print results
    print(f'Wilcoxon Signed-Rank Test between models:\nStatistical Test:{statistic:.4f},P-value:{p_value:.4f}')
    print(f'Mean difference: {mean_difference:.4f}')
    
    if p_value < 0.05:
        if mean_difference > 0:
            print("Model 1 is performing better than Model 2.")
            print()
        else:
            print("Model 2 is performing better than Model 1.")
            print()
    else:
        print("There is no statistically significant difference between the two models.")
        print()
    
    return statistic, p_value, mean_difference


# Statistical testing for comparison of performance of models
wilcoxon_signed_rank_test(model[0], model[1], breast_x_test, y_test)  # Logistic Regression vs Decision Tree
wilcoxon_signed_rank_test(model[0], model[2], breast_x_test, y_test)  # Logistic Regression vs Random Forest
wilcoxon_signed_rank_test(model[0], model[3], breast_x_test, y_test)  # Logistic Regression vs SVC
wilcoxon_signed_rank_test(model[1], model[2], breast_x_test, y_test)  # Decision Tree vs Random Forest
wilcoxon_signed_rank_test(model[1], model[3], breast_x_test, y_test)  # Decision Tree vs SVC
wilcoxon_signed_rank_test(model[2], model[3], breast_x_test, y_test)  # Random Forest vs SVC




# ###########Begining of Lung Cancer Prediction##############

print()
print("*****Lung Cancer Prediction*****")
print()

lung_df = pd.read_csv("H:\\Users\\SAHIL\\Downloads\\survey lung cancer.csv")

# HERE 1 stands fror NO & 2 stands for YES
lung_df.head()


lung_df.shape

lung_df.info()

lung_df.describe()

lung_df.isnull().sum()

lung_df.isna().sum()

lung_df['LUNG_CANCER'] = lung_df['LUNG_CANCER'].astype('category')
lung_df.dtypes

#creating object of LabelEncoder 
le_y = LabelEncoder()
lung_df['GENDER'] = le_y.fit_transform(lung_df['GENDER'].values)

lung_df['LUNG_CANCER'] = le_y.fit_transform(lung_df['LUNG_CANCER'].values)

lung_df.dtypes

lung_df.head()
# In GENDER attribute: 1 for M & 0 for F
# In LUNG_CANCER attribute: 1 for YES & 0 for NO

lung_df['LUNG_CANCER'].value_counts() 

type(lung_df)

# lung_df.groupby('LUNG_CANCER').mean()

lung_x = lung_df.drop('LUNG_CANCER', axis = 1)
lung_y = np.array(lung_df['LUNG_CANCER'])

#appling the feature selection model to the dataset
etc = ExtraTreesClassifier()
etc.fit(lung_x, lung_y)


print(etc.feature_importances_)

feature_importance = pd.Series(etc.feature_importances_, index = lung_x.columns)
feature_importance.nlargest(18).plot(kind = 'barh')
plt.show()



lung_feat_corr =lung_df[['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']].corr()
plt.figure(figsize = (15,10))
sns.heatmap(lung_feat_corr, annot = True)

lung_feat =lung_df[['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']]

# observinng outliers
plt.figure(figsize = (12,12))

for i , col in enumerate(lung_feat.columns[:12]):
    plt.subplot(4,4, i+1)
    sns.boxplot(x =  col, data =lung_feat)
    
plt.show()

plt.figure(figsize = (12,12))
for i , col in enumerate(lung_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.histplot(x =  col, data = lung_feat, kde= True)
    
plt.show()

from scipy import stats
plt.figure(figsize = (12,12))
for i , col in enumerate(lung_feat.columns[:12]):
    plt.subplot(6,2,i+1)
    stats.probplot(lung_feat[col], dist = 'norm', plot = plt)
    
plt.show()

new_lung_feat = lung_feat.copy()
lung_clean_df = outliers(new_lung_feat)


plt.figure(figsize = (12,12))

for i , col in enumerate(lung_clean_df.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =lung_clean_df)
    
plt.show()

x_train, x_test, y_train, y_test = train_test_split(lung_clean_df, lung_y, test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

lung_x_train = sc.fit_transform(x_train)
lung_x_test = sc.fit_transform(x_test)

#create a function for the models
def models(lung_x_train, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log_lung = LogisticRegression(random_state = 0)
    log_lung.fit(lung_x_train, y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree_lung = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree_lung.fit(lung_x_train, y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest_lung = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state = 0)
    forest_lung.fit(lung_x_train, y_train)

    #Support Vector Classifier
    from sklearn.svm import SVC
    svc_lung = SVC(kernel = 'rbf', random_state = 0, probability = True)
    svc_lung.fit(lung_x_train, y_train)

    # print the model accuracy on the training data
    print('[0] Logistic Regression Training Accuracy: ', log_lung.score(lung_x_train , y_train))
    print('[1] Decision Tree Classifier Training Accuracy: ', tree_lung.score(lung_x_train , y_train))
    print('[2] Random Forest Classifier Training Accuracy: ', forest_lung.score(lung_x_train , y_train))
    print('[3] Support Vector Classifier Training Accuracy: ', svc_lung.score(lung_x_train , y_train))

    return log_lung, tree_lung, forest_lung, svc_lung

model = models(lung_x_train, y_train)



# Evaluate models with cross-validation
cross_validation_evaluation(model,lung_clean_df , lung_y)
#test model accuracy on test data on cofusion matrix
from sklearn. metrics import confusion_matrix

print('0 - Logistic Regression \n1 - Decision Tree Classifier \n2 - Random Forest Classifier \n3 - SVC')

for i in range(len(model)):
    print(model[i])
    cm = confusion_matrix(y_test, model[i].predict(lung_x_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]

    print(cm)
    print('Testing Accuracy = ', (TP+TN)/(TP+TN+FP+FN))
    print()
    if TN + FP == 0:
        print('0')
    else:
         print('Specificity - ',TN / (TN + FP))


from sklearn.metrics import roc_curve, auc

lung_log_pred = model[0].decision_function(lung_x_test)
lung_dcn_tree_pred = model[1].predict_proba(lung_x_test)[:,1]
lung_rndm_forest_pred = model[2].predict_proba(lung_x_test)[:,1]
lung_svc_pred =  model[3].decision_function(lung_x_test)

lung_log_fpr, lung_log_tpr, threshold = roc_curve(y_test, lung_log_pred)
lung_auc_log = auc(lung_log_fpr, lung_log_tpr)

lung_svc_fpr, lung_svc_tpr, threshold = roc_curve(y_test, lung_svc_pred)
lung_auc_svc = auc(lung_svc_fpr, lung_svc_tpr)

lung_decision_tree_fpr, lung_decision_tree_tpr, threshold = roc_curve(y_test, lung_dcn_tree_pred)
lung_auc_dcn_tree = auc(lung_decision_tree_fpr, lung_decision_tree_tpr)

lung_rndm_forest_fpr, lung_rndm_forest_tpr, threshold = roc_curve(y_test, lung_rndm_forest_pred)
lung_auc_rndm_forest = auc(lung_rndm_forest_fpr, lung_rndm_forest_tpr)


print(lung_auc_log)
print(lung_auc_dcn_tree)
print(lung_auc_rndm_forest)
print(lung_auc_svc)

plt.figure(figsize = (5,5), dpi = 100)
plt.plot(lung_log_fpr, lung_log_tpr, linestyle = '-', label = 'Logistic Regression (auc = %0.3f)' %lung_auc_log)

plt.plot(lung_decision_tree_fpr, lung_decision_tree_tpr, linestyle= '-', label = 'Decision Tree (auc = %0.3f)' %lung_auc_dcn_tree)

plt.plot(lung_rndm_forest_fpr, lung_rndm_forest_tpr, linestyle = '-', label = 'Random Forest Classifier (auc = %0.3f)' %lung_auc_rndm_forest)

plt.plot(lung_svc_fpr, lung_svc_tpr, linestyle = '-', label = 'SVC (auc = %0.3f)' %lung_auc_svc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()

#classification report 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print(model[i])
    print(classification_report(y_test, model[i].predict(lung_x_test)))
    print(accuracy_score(y_test, model[i].predict(lung_x_test)))
    print()


# Statistical testing for comparison of performance of models
wilcoxon_signed_rank_test(model[0], model[1], lung_x_test, y_test)  # Logistic Regression vs Decision Tree
wilcoxon_signed_rank_test(model[0], model[2], lung_x_test, y_test)  # Logistic Regression vs Random Forest
wilcoxon_signed_rank_test(model[0], model[3], lung_x_test, y_test)  # Logistic Regression vs SVC
wilcoxon_signed_rank_test(model[1], model[2], lung_x_test, y_test)  # Decision Tree vs Random Forest
wilcoxon_signed_rank_test(model[1], model[3], lung_x_test, y_test)  # Decision Tree vs SVC
wilcoxon_signed_rank_test(model[2], model[3], lung_x_test, y_test)  # Random Forest vs SVC

# ######################End of Lung cancer predition########################

# ######################Begining of Chronic Kidney Disease Prediction##################

print()
print("*****Chronic Kidney Disease Prediction*****")
print()

kidney_df = pd.read_csv("H:\\Users\\SAHIL\\Downloads\\kidney disease.csv")

kidney_df.head()

kidney_df.shape

kidney_df.info()

kidney_df.describe()

kidney_df.isnull().sum()

kidney_df.isna().sum()

kidney_df.duplicated().sum()

kidney_df['Class'].value_counts() 

kidney_df.groupby('Class').mean()

# sns.pairplot(df.iloc[:,1:6], hue = 'diagnosis')

kidney_x = kidney_df.drop('Class', axis = 1)
kidney_y = np.array(kidney_df['Class'])

#appling the feature selection model to the dataset
etc = ExtraTreesClassifier()
etc.fit(kidney_x, kidney_y)


print(etc.feature_importances_)

feature_importance = pd.Series(etc.feature_importances_, index = kidney_x.columns)
feature_importance.nlargest(12).plot(kind = 'barh')
plt.show()

kid_feat_corr =kidney_df[['Wbcc', 'Bp','Bu','Su','Sod','Rbc','Sc','Rbcc','Hemo', 'Al','Htn','Sg','Class']].corr()
plt.figure(figsize = (15,10))
sns.heatmap(kid_feat_corr, annot = True)

kidney_feat =kidney_df[['Wbcc', 'Bp','Bu','Su','Sod','Rbc','Sc','Rbcc','Hemo', 'Al','Htn','Sg','Class']]

# observinng outliers
plt.figure(figsize = (12,12))

for i , col in enumerate(kidney_feat.columns[:12]):
    plt.subplot(4,4, i+1)
    sns.boxplot(x =  col, data =kidney_feat)
    
plt.show()

plt.figure(figsize = (12,12))
for i , col in enumerate(kidney_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.histplot(x =  col, data = kidney_feat, kde= True)
    
plt.show()

from scipy import stats
plt.figure(figsize = (12,12))
for i , col in enumerate(kidney_feat.columns[:12]):
    plt.subplot(6,2,i+1)
    stats.probplot(kidney_feat[col], dist = 'norm', plot = plt)
    
plt.show()

new_kidney_feat = kidney_feat.copy()
kid_clean_df = outliers(new_kidney_feat)

plt.figure(figsize = (12,12))

for i , col in enumerate(kid_clean_df.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =kid_clean_df)
    
plt.show()

x_train, x_test, y_train, y_test = train_test_split(kid_clean_df, kidney_y, test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

kid_x_train_sc = sc.fit_transform(x_train)
kid_x_test_sc = sc.fit_transform(x_test)

#create a function for the models
def models(kid_x_train_sc, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state = 0)
    log.fit(kid_x_train_sc, y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree.fit(kid_x_train_sc, y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state = 0)
    forest.fit(kid_x_train_sc, y_train)

    #Support Vector Classifier
    from sklearn.svm import SVC
    svc = SVC(kernel = 'rbf', random_state = 0, probability = True)
    svc.fit(kid_x_train_sc, y_train)

    # print the model accuracy on the training data
    print('[0] Logistic Regression Training Accuracy: ', log.score(kid_x_train_sc , y_train))
    print('[1] Decision Tree Classifier Training Accuracy: ', tree.score(kid_x_train_sc , y_train))
    print('[2] Random Forest Classifier Training Accuracy: ', forest.score(kid_x_train_sc , y_train))
    print('[3] Support Vector Classifier Training Accuracy: ', svc.score(kid_x_train_sc , y_train))

    return log, tree, forest, svc

model = models(kid_x_train_sc, y_train)


# Evaluate models with cross-validation
cross_validation_evaluation(model,kid_clean_df , kidney_y)
#test model accuracy on test data on cofusion matrix
from sklearn. metrics import confusion_matrix

print('0 - Logistic Regression \n1 - Decision Tree Classifier \n2 - Random Forest Classifier \n3 - SVC')

for i in range(len(model)):
    print(model[i])
    cm = confusion_matrix(y_test, model[i].predict(kid_x_test_sc))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]

    print(cm)
    print('Testing Accuracy = ', (TP+TN)/(TP+TN+FP+FN))
    print()

    if TN + FP == 0:
        print('0')
    else:
         print('Specificity - ',TN / (TN + FP))

from sklearn.metrics import roc_curve, auc

log_pred = model[0].decision_function(kid_x_test_sc)
dcn_tree_pred = model[1].predict_proba(kid_x_test_sc)[:,1]
rndm_forest_pred = model[2].predict_proba(kid_x_test_sc)[:,1]
svc_pred =  model[3].decision_function(kid_x_test_sc)

log_fpr, log_tpr, threshold = roc_curve(y_test, log_pred)
auc_log = auc(log_fpr, log_tpr)

svc_fpr, svc_tpr, threshold = roc_curve(y_test, svc_pred)
auc_svc = auc(svc_fpr, svc_tpr)

decision_tree_fpr, decision_tree_tpr, threshold = roc_curve(y_test, dcn_tree_pred)
auc_dcn_tree = auc(decision_tree_fpr, decision_tree_tpr)

rndm_forest_fpr, rndm_forest_tpr, threshold = roc_curve(y_test, rndm_forest_pred)
auc_rndm_forest = auc(rndm_forest_fpr, rndm_forest_tpr)


print(auc_log)
print(auc_dcn_tree)
print(auc_rndm_forest)
print(auc_svc)

plt.figure(figsize = (5,5), dpi = 100)
plt.plot(log_fpr, log_tpr, linestyle = '-', label = 'Logistic Regression (auc = %0.3f)' %auc_log)

plt.plot(decision_tree_fpr, decision_tree_tpr, linestyle= '-', label = 'Decision Tree (auc = %0.3f)' %auc_dcn_tree)

plt.plot(rndm_forest_fpr, rndm_forest_tpr, linestyle = '-', label = 'Random Forest Classifier (auc = %0.3f)' %auc_rndm_forest)

plt.plot(svc_fpr, svc_tpr, linestyle = '-', label = 'SVC (auc = %0.3f)' %auc_svc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()

#classification report 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print(model[i])
    print(classification_report(y_test, model[i].predict(kid_x_test_sc)))
    print(accuracy_score(y_test, model[i].predict(kid_x_test_sc)))
    print()


# Statistical testing for comparison of performance of models
wilcoxon_signed_rank_test(model[0], model[1], kid_x_test_sc, y_test)  # Logistic Regression vs Decision Tree
wilcoxon_signed_rank_test(model[0], model[2], kid_x_test_sc, y_test)  # Logistic Regression vs Random Forest
wilcoxon_signed_rank_test(model[0], model[3], kid_x_test_sc, y_test)  # Logistic Regression vs SVC
wilcoxon_signed_rank_test(model[1], model[2], kid_x_test_sc, y_test)  # Decision Tree vs Random Forest
wilcoxon_signed_rank_test(model[1], model[3], kid_x_test_sc, y_test)  # Decision Tree vs SVC
wilcoxon_signed_rank_test(model[2], model[3], kid_x_test_sc, y_test)  # Random Forest vs SVC
# ######################End of Chronic Kidney Disease predition########################


# ######################Begining of Heart Attack Prediction##################

print()
print("*****Heart Attack Prediction*****")
print()
heart_df = pd.read_csv("H:\\Users\\SAHIL\\Downloads\\Heart Attack.csv")
heart_df.head(10)

heart_df.shape

heart_df.info()

heart_df.describe()

heart_df.isnull().sum()

# df1.rename(columns={'Outcome': 'class'}, inplace=True)

heart_df['class'].value_counts() 

heart_df['class'].dtypes

print(heart_df.columns)

#creating object of LabelEncoder 
le_y = LabelEncoder()
heart_df['class'] = le_y.fit_transform(heart_df['class'].values)

heart_df['class']

heart_df.groupby('class').mean()

# sns.pairplot(heart_df ,hue = 'class')

heart_x = heart_df.drop('class', axis = 1)
heart_y = np.array(heart_df['class'])

#appling the feature selection model to the dataset
etc = ExtraTreesClassifier()
etc.fit(heart_x, heart_y)


print(etc.feature_importances_)

feature_importance = pd.Series(etc.feature_importances_, index = heart_x.columns)
feature_importance.nlargest(10).plot(kind = 'barh')
plt.show()

heart_feat_corr =heart_df[['troponin','kcm', 'age', 'glucose', 'pressurelow', 'pressurehight', 'impluse']].corr()
plt.figure(figsize = (15,10))
sns.heatmap(heart_feat_corr, annot = True)

heart_feat = heart_df[['troponin','kcm', 'age', 'glucose', 'pressurelow', 'pressurehight', 'impluse']]

plt.figure(figsize = (12,12))

for i , col in enumerate(heart_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =heart_feat)
    
plt.show()

plt.figure(figsize = (12,12))
for i , col in enumerate(heart_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.histplot(x =  col, data = heart_feat, kde= True)
    
plt.show()

from scipy import stats
plt.figure(figsize = (12,12))
for i , col in enumerate(heart_feat.columns[:12]):
    plt.subplot(6,2,i+1)
    stats.probplot(heart_feat[col], dist = 'norm', plot = plt)
    
plt.show()

new_heart_feat = heart_feat.copy()
heart_clean_df = outliers(new_heart_feat)


plt.figure(figsize = (12,12))

for i , col in enumerate(heart_clean_df.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =heart_clean_df)
    
plt.show()



x_train, x_test, y_train, y_test = train_test_split(heart_clean_df, heart_y , test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

heart_x_train = sc.fit_transform(x_train)
heart_x_test = sc.fit_transform(x_test)

#create a function for the models
def models(heart_x_train, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    heart_log = LogisticRegression(random_state = 0)
    heart_log.fit(heart_x_train, y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    heart_tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    heart_tree.fit(heart_x_train, y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    heart_forest = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state = 0)
    heart_forest.fit(heart_x_train, y_train)

    #Support Vector Clasheart
    from sklearn.svm import SVC
    heart_svc = SVC(kernel = 'rbf', random_state = 0, probability = True)
    heart_svc.fit(heart_x_train, y_train)

    # print the model accuracy on the training data
    print('[0] Logistic Regression Training Accuracy: ', heart_log.score(heart_x_train , y_train))
    print('[1] Decision Tree Classifier Training Accuracy: ', heart_tree.score(heart_x_train , y_train))
    print('[2] Random Forest Classifier Training Accuracy: ', heart_forest.score(heart_x_train , y_train))
    print('[3] Support Vector Classifier Training Accuracy: ', heart_svc.score(heart_x_train , y_train))

    return heart_log, heart_tree, heart_forest, heart_svc

model = models(heart_x_train, y_train)


# Evaluate models with cross-validation
cross_validation_evaluation(model,heart_clean_df , heart_y)
#test model accuracy on test data on cofusion matrix
from sklearn. metrics import confusion_matrix

print('0 - Logistic Regression \n1 - Decision Tree Classifier \n2 - Random Forest Classifier \n3 - SVC')

for i in range(len(model)):
    print(model[i])
    cm = confusion_matrix(y_test, model[i].predict(heart_x_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]

    print(cm)
    print('Testing Accuracy = ', (TP+TN)/(TP+TN+FP+FN))
    print()

    if TN + FP == 0:
        print('0')
    else:
         print('Specificity - ',TN / (TN + FP))


from sklearn.metrics import roc_curve, auc

heart_log_pred = model[0].decision_function(heart_x_test)
heart_dcn_tree_pred = model[1].predict_proba(heart_x_test)[:,1]
heart_rndm_forest_pred = model[2].predict_proba(heart_x_test)[:,1]
heart_svc_pred =  model[3].decision_function(heart_x_test)

heart_log_fpr, heart_log_tpr, threshold = roc_curve(y_test, heart_log_pred)
heart_auc_log = auc(heart_log_fpr, heart_log_tpr)

heart_svc_fpr, heart_svc_tpr, threshold = roc_curve(y_test, heart_svc_pred)
heart_auc_svc = auc(heart_svc_fpr, heart_svc_tpr)

heart_decision_tree_fpr, heart_decision_tree_tpr, threshold = roc_curve(y_test, heart_dcn_tree_pred)
heart_auc_dcn_tree = auc(heart_decision_tree_fpr, heart_decision_tree_tpr)

heart_rndm_forest_fpr, heart_rndm_forest_tpr, threshold = roc_curve(y_test, heart_rndm_forest_pred)
heart_auc_rndm_forest = auc(heart_rndm_forest_fpr, heart_rndm_forest_tpr)


print(heart_auc_log)
print(heart_auc_dcn_tree)
print(heart_auc_rndm_forest)
print(heart_auc_svc)

plt.figure(figsize = (5,5), dpi = 100)
plt.plot(heart_log_fpr, heart_log_tpr, linestyle = '-', label = 'Logistic Regression (auc = %0.3f)' %heart_auc_log)
plt.plot(heart_svc_fpr, heart_svc_tpr, linestyle = '-', label = 'SVC (auc = %0.3f)' %heart_auc_svc)
plt.plot(heart_decision_tree_fpr, heart_decision_tree_tpr, linestyle= '-', label = 'Decision Tree (auc = %0.3f)' %heart_auc_dcn_tree)
plt.plot(heart_rndm_forest_fpr, heart_rndm_forest_tpr, linestyle = '-', label = 'Random Forest Classifier (auc = %0.3f)' %heart_auc_rndm_forest)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()


#classification report 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print(model[i])
    print(classification_report(y_test, model[i].predict(heart_x_test)))
    print(accuracy_score(y_test, model[i].predict(heart_x_test)))
    print()


# Statistical testing for comparison of performance of models
wilcoxon_signed_rank_test(model[0], model[1], heart_x_test, y_test)  # Logistic Regression vs Decision Tree
wilcoxon_signed_rank_test(model[0], model[2], heart_x_test, y_test)  # Logistic Regression vs Random Forest
wilcoxon_signed_rank_test(model[0], model[3], heart_x_test, y_test)  # Logistic Regression vs SVC
wilcoxon_signed_rank_test(model[1], model[2], heart_x_test, y_test)  # Decision Tree vs Random Forest
wilcoxon_signed_rank_test(model[1], model[3], heart_x_test, y_test)  # Decision Tree vs SVC
wilcoxon_signed_rank_test(model[2], model[3], heart_x_test, y_test)  # Random Forest vs SVC

# ######################End of Heart Attack Prediction########################


# ######################Begining of Diabetes Prediction##################


print()
print("*****Diabetes Disease Diagnosis*****")
print()
dib_df = pd.read_csv("C:\\Nishu\\diabetes.csv")
dib_df.head(10)

dib_df.shape

dib_df.info()

dib_df.describe()

dib_df.isnull().sum()

# df1.rename(columns={'Outcome': 'class'}, inplace=True)

dib_df['class'].value_counts() 

dib_df['class'].dtypes

print(dib_df.columns)

#creating object of LabelEncoder 
le_y = LabelEncoder()
dib_df['class'] = le_y.fit_transform(dib_df['class'].values)

dib_df['class']

dib_df.groupby('class').mean()

# sns.pairplot(heart_df ,hue = 'class')

dib_x = dib_df.drop('class', axis = 1)
dib_y = np.array(dib_df['class'])

#appling the feature selection model to the dataset
etc = ExtraTreesClassifier()
etc.fit(dib_x, dib_y)


print(etc.feature_importances_)

feature_importance = pd.Series(etc.feature_importances_, index = dib_x.columns)
feature_importance.nlargest(10).plot(kind = 'barh')
plt.show()

dib_feat_corr =dib_df[['insu', 'skin','pres','preg','pedi','mass','age','plasma']].corr()
plt.figure(figsize = (15,10))
sns.heatmap(dib_feat_corr, annot = True)

dib_feat = dib_df[['insu', 'skin','pres','preg','pedi','mass','age','plasma']]

plt.figure(figsize = (12,12))

for i , col in enumerate(dib_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =dib_feat)
    
plt.show()

plt.figure(figsize = (12,12))
for i , col in enumerate(dib_feat.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.histplot(x =  col, data = dib_feat, kde= True)
    
plt.show()

from scipy import stats
plt.figure(figsize = (12,12))
for i , col in enumerate(dib_feat.columns[:12]):
    plt.subplot(6,2,i+1)
    stats.probplot(dib_feat[col], dist = 'norm', plot = plt)
    
plt.show()

new_dib_feat = dib_feat.copy()
dib_clean_df = outliers(new_dib_feat)


plt.figure(figsize = (12,12))

for i , col in enumerate(dib_clean_df.columns[:12]):
    plt.subplot(3,4, i+1)
    sns.boxplot(x =  col, data =dib_clean_df)
    
plt.show()



x_train, x_test, y_train, y_test = train_test_split(dib_clean_df, dib_y , test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

dib_x_train = sc.fit_transform(x_train)
dib_x_test = sc.fit_transform(x_test)

#create a function for the models
def models(dib_x_train, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    dib_log = LogisticRegression(random_state = 0)
    dib_log.fit(dib_x_train, y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    dib_tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    dib_tree.fit(dib_x_train, y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    dib_forest = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state = 0)
    dib_forest.fit(dib_x_train, y_train)

    #Support Vector Clasheart
    from sklearn.svm import SVC
    dib_svc = SVC(kernel = 'rbf', random_state = 0, probability = True)
    dib_svc.fit(dib_x_train, y_train)

    # print the model accuracy on the training data
    print('[0] Logistic Regression Training Accuracy: ', dib_log.score(dib_x_train , y_train))
    print('[1] Decision Tree Classifier Training Accuracy: ', dib_tree.score(dib_x_train , y_train))
    print('[2] Random Forest Classifier Training Accuracy: ', dib_forest.score(dib_x_train , y_train))
    print('[3] Support Vector Classifier Training Accuracy: ', dib_svc.score(dib_x_train , y_train))

    return dib_log, dib_tree, dib_forest, dib_svc

model = models(dib_x_train, y_train)


# Evaluate models with cross-validation
cross_validation_evaluation(model,dib_clean_df , dib_y)
#test model accuracy on test data on cofusion matrix
from sklearn. metrics import confusion_matrix

print('0 - Logistic Regression \n1 - Decision Tree Classifier \n2 - Random Forest Classifier \n3 - SVC')

for i in range(len(model)):
    print(model[i])
    cm = confusion_matrix(y_test, model[i].predict(dib_x_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]

    print(cm)
    print('Testing Accuracy = ', (TP+TN)/(TP+TN+FP+FN))
    print()

    if TN + FP == 0:
        print('0')
    else:
         print('Specificity - ',TN / (TN + FP))


from sklearn.metrics import roc_curve, auc

dib_log_pred = model[0].decision_function(dib_x_test)
dib_dcn_tree_pred = model[1].predict_proba(dib_x_test)[:,1]
dib_rndm_forest_pred = model[2].predict_proba(dib_x_test)[:,1]
dib_svc_pred =  model[3].decision_function(dib_x_test)

dib_log_fpr, dib_log_tpr, threshold = roc_curve(y_test, dib_log_pred)
dib_auc_log = auc(dib_log_fpr, dib_log_tpr)

dib_svc_fpr, dib_svc_tpr, threshold = roc_curve(y_test, dib_svc_pred)
dib_auc_svc = auc(dib_svc_fpr, dib_svc_tpr)

dib_decision_tree_fpr, dib_decision_tree_tpr, threshold = roc_curve(y_test, dib_dcn_tree_pred)
dib_auc_dcn_tree = auc(dib_decision_tree_fpr, dib_decision_tree_tpr)

dib_rndm_forest_fpr, dib_rndm_forest_tpr, threshold = roc_curve(y_test, dib_rndm_forest_pred)
dib_auc_rndm_forest = auc(dib_rndm_forest_fpr, dib_rndm_forest_tpr)


print(dib_auc_log)
print(dib_auc_dcn_tree)
print(dib_auc_rndm_forest)
print(dib_auc_svc)

plt.figure(figsize = (5,5), dpi = 100)
plt.plot(dib_log_fpr, dib_log_tpr, linestyle = '-', label = 'Logistic Regression (auc = %0.3f)' %dib_auc_log)
plt.plot(dib_svc_fpr, dib_svc_tpr, linestyle = '-', label = 'SVC (auc = %0.3f)' %dib_auc_svc)
plt.plot(dib_decision_tree_fpr, dib_decision_tree_tpr, linestyle= '-', label = 'Decision Tree (auc = %0.3f)' %dib_auc_dcn_tree)
plt.plot(dib_rndm_forest_fpr, dib_rndm_forest_tpr, linestyle = '-', label = 'Random Forest Classifier (auc = %0.3f)' %dib_auc_rndm_forest)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()


#classification report 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print(model[i])
    print(classification_report(y_test, model[i].predict(dib_x_test)))
    print(accuracy_score(y_test, model[i].predict(dib_x_test)))
    print()


# Statistical testing for comparison of performance of models
wilcoxon_signed_rank_test(model[0], model[1], dib_x_test, y_test)  # Logistic Regression vs Decision Tree
wilcoxon_signed_rank_test(model[0], model[2], dib_x_test, y_test)  # Logistic Regression vs Random Forest
wilcoxon_signed_rank_test(model[0], model[3], dib_x_test, y_test)  # Logistic Regression vs SVC
wilcoxon_signed_rank_test(model[1], model[2], dib_x_test, y_test)  # Decision Tree vs Random Forest
wilcoxon_signed_rank_test(model[1], model[3], dib_x_test, y_test)  # Decision Tree vs SVC
wilcoxon_signed_rank_test(model[2], model[3], dib_x_test, y_test)  # Random Forest vs SVC

# ######################End of Diabetes Prediction ########################
```
